```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"https://gist.githubusercontent.com/Pshemkoo/1dcd2f7d85ea7019d394391fcd4450c0/raw/9f9d38c148ee8011569b44678906011872eb9fc5/gistfile1.txt") # real
data[data['years']==3]
```

```python
data.info()
```

```python
data.describe()
```

```python
X = data["years"]
y = data["salary"]
```

```python
plt.figure(figsize=(1,1))
sns.pairplot(data,x_vars=["years"],y_vars=["salary"],size=5,kind="scatter")
plt.xlabel("years")
plt.ylabel("salary")
plt.title("years & salary" )
plt.show()
```

```python
X = data["years"]
y = data["salary"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
x_train , x_test , y_tarin , y_test = train_test_split(X, y , train_size=0.8 , random_state = 42)
model = LinearRegression()
model.fit(x_train,y_tarin)
```

```python
model.coef_ ; model.intercept_
```

```python
y_pred = model.predict(x_test.values.reshape(-1,1))
```

```python
xin = int(input("entr year :"))
xin = [[xin]]
y_pred_x = model.predict(xin)
y_pred_x
```

```python
MSE = mean_squared_error(y_test,y_pred)
print("MSE = ",MSE)
MAE = mean_absolute_error(y_test,y_pred)
print("MAE = ",MAE)
r2_score = r2_score(y_test,y_pred)
print("r2_score = ",r2_score)
```

```python
ys = pd.DataFrame()
ys['y_test'] = y_test
ys['y_pred'] = y_pred
ys['cost'] = y_test - y_pred
ys
```

```python
# Plotting the actual and predicted values

plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_test, y_pred, color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# plotting the error
c = [i for i in range(1,len(y_test)+1)]
plt.plot(c,y_test-y_pred,color='green',linestyle='-')
plt.xlabel('index')
plt.ylabel('Error')
plt.title('Error Value')
plt.show()
```

```python
# Plotting the actual and predicted values

c = [i for i in range(1,len(y_test)+1)]
plt.plot(c,y_test,color='r',linestyle='-')
plt.plot(c,y_pred,color='b',linestyle='-')
plt.legend(['y_test','y_pred'])
plt.xlabel('Salary')
plt.ylabel('index')
plt.title('Prediction')
plt.show()
```
